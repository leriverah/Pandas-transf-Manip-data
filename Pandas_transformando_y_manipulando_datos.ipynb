# **Problema de Negocio**

```python
import pandas as pd
```

```python
datos = pd.read_json('datos_hosting.json')
datos.head()
```

```python
datos = pd.json_normalize(datos ['info_inmuebles'])
datos.head()
```

## **Retirando las listas**

```python
# Consultando el nombre de las columnas

columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[3:])
datos
```

```python
# inplace = True - ejecutar en el mismo dataframe
# drop = True - borrar memoria

datos.reset_index(inplace = True, drop = True)
datos.head()
```
